### Helper function to load PDF documents

### 1. Helper function for basic text cleaning

````python
def clean_text(text: str) -> str:
    """
    Performs basic text cleaning:
    - Lowercasing
    - Removal of consecutive tabs
    - Replacement of whitespace sequences with a single space
    """
    text = text.lower() # Lowercase the text
    text = re.sub(r'\t+', ' ', text) # Replace consecutive tabs with a single space
    text = re.sub(r'\s+', ' ', text) # Replace any sequence of whitespace characters with a single space
    return text.strip()

# Example usage:
# cleaned_example = clean_text("  This\t is   a  TEST\ttext with   various\twhitespaces. ")
# print(f"Original: '  This\t is   a  TEST\ttext with   various\twhitespaces. '")
# print(f"Cleaned: '{cleaned_example}'")
````

### 2. Helper function for splitting documents into chunks

````python
def split_docs(documents: list, chunk_size: int = 1000, chunk_overlap: int = 200):
    """
    Splits a list of documents into smaller chunks.

    Args:
        documents (list): A list of langchain Document objects.
        chunk_size (int): The maximum size of each chunk.
        chunk_overlap (int): The number of characters to overlap between chunks.

    Returns:
        list: A list of langchain Document objects representing the chunks.
    """
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=chunk_size,
        chunk_overlap=chunk_overlap,
        length_function=len,
        is_separator_regex=False,
    )
    chunks = text_splitter.split_documents(documents)
    print(f"Split {len(documents)} documents into {len(chunks)} chunks.")
    return chunks

# Note: This function expects `documents` to be a list of Langchain Document objects.
````

### 3. Helper function to create embeddings

````python
def create_embeddings(chunks: list):
    """
    Creates embeddings for a list of text chunks using OpenAI's 'text-embedding-3-small' model.

    Args:
        chunks (list): A list of langchain Document objects, where each object's page_content is a chunk.

    Returns:
        list: A list of numpy arrays, where each array is the embedding for a corresponding chunk.
    """
    texts = [chunk.page_content for chunk in chunks]
    print(f"Creating embeddings for {len(texts)} chunks...")
    response = client.embeddings.create(input=texts, model="text-embedding-3-small")
    embeddings = [np.array(embedding_data.embedding) for embedding_data in response.data]
    print("Embeddings created.")
    return embeddings
````

### 4. Helper function to build a vector database (FAISS)

````python
def build_vector_db(chunks: list, embeddings: list):
    """
    Builds a FAISS IndexFlatL2 vector database from embeddings and associates them with chunks.

    Args:
        chunks (list): A list of langchain Document objects.
        embeddings (list): A list of numpy arrays, corresponding to the embeddings of the chunks.

    Returns:
        tuple: A tuple containing the FAISS index and the list of chunks.
    """
    if not embeddings:
        print("No embeddings to build the vector database.")
        return None, []

    # Ensure embeddings are in a 2D array format for FAISS
    embeddings_array = np.array(embeddings).astype('float32')
    dimension = embeddings_array.shape[1]

    index = faiss.IndexFlatL2(dimension) # L2 distance for similarity
    index.add(embeddings_array)

    print(f"FAISS index built with {index.ntotal} vectors.")
    return index, chunks
````

### 5. Helper function to search for similar chunks

````python
def search_chunks(query: str, index, chunks: list, k: int = 3):
    """
    Searches the FAISS index for relevant chunks based on a query.

    Args:
        query (str): The user's query string.
        index: The FAISS index.
        chunks (list): The list of langchain Document objects corresponding to the index.
        k (int): The number of top-k relevant chunks to retrieve.

    Returns:
        list: A list of the most relevant langchain Document objects.
    """
    query_embedding_response = client.embeddings.create(input=[query], model="text-embedding-3-small")
    query_embedding = np.array(query_embedding_response.data[0].embedding).astype('float32').reshape(1, -1)

    distances, indices = index.search(query_embedding, k) # Search for top-k similar vectors

    relevant_chunks = [chunks[i] for i in indices[0]]
    print(f"Found {len(relevant_chunks)} relevant chunks for the query.")
    return relevant_chunks
````

### 6. Function to generate answer using OpenAI GPT-4o-mini

````python
def _raw_generate_answer(query, context_chunks):
    """
    Generates an answer to a query using the provided context and OpenAI's GPT-4o-mini.
    This is the core generation logic without critics.

    Args:
        query (str): The user's question.
        context_chunks (list): A list of langchain Document objects to use as context.

    Returns:
        str: The generated answer.
    """
    # Combine the text from the retrieved Documents
    context = "\n\n".join(doc.page_content for doc in context_chunks)

    response = client.chat.completions.create(
        model = "gpt-4o-mini",
        messages=[
            {"role": "system", "content": "Answer based on the context provided. If the answer is not available in the context, politely state that you cannot provide an answer based on the given information."},
            {"role": "user", "content": f"Context:\n{context}\n\nQuestion: {query}"}
        ],
        temperature=0.2
    )
    return response.choices[0].message.content

def generate_answer(query: str, index, chunks):
    """
    Generates an answer to a query using the Self-RAG system (retrieval and generation critics).

    Args:
        query (str): The user's question.
        index: The FAISS index for retrieval.
        chunks (list): The list of langchain Document objects corresponding to the index.

    Returns:
        tuple: A tuple containing the generated answer (str), the relevant chunks (list),
               and the generation critic's assessment (dict).
    """
    # 1. Search for relevant chunks
    retrieved_chunks = search_chunks(query, index, chunks)

    # 2. Apply Retrieval Critic
    relevant_chunks = implement_retrieval_critic(query, retrieved_chunks)

    if not relevant_chunks:
        print("Self-RAG: Retrieval Critic found no suitable context. Skipping answer generation.")
        return "No suitable context found to answer the question.", [], {"Faithful": "N/A", "Relevant": "N/A", "Explanation": "No context."}

    # 3. Generate answer using _raw_generate_answer
    generated_answer = _raw_generate_answer(query, relevant_chunks)

    # 4. Apply Generation Critic
    generation_critic_result = implement_generation_critic(query, relevant_chunks, generated_answer)

    return generated_answer, relevant_chunks, generation_critic_result

print("Refactored generate_answer to include Self-RAG critics.")
````

## Implement Retrieval Critic Function

````python
import json # Ensure json is imported for parsing LLM response

def implement_retrieval_critic(query: str, retrieved_chunks: list):
    """
    Evaluates the relevance and utility of retrieved document chunks using an LLM critic.

    Args:
        query (str): The user's query.
        retrieved_chunks (list): A list of langchain Document objects.

    Returns:
        list: The list of relevant langchain Document objects if deemed suitable, otherwise an empty list.
    """
    if not retrieved_chunks:
        print("Retrieval Critic: No chunks to evaluate. Returning empty list.")
        return []

    # Prepare chunks for the prompt
    context_chunks_str = json.dumps([doc.page_content for doc in retrieved_chunks], indent=2)

    retrieval_critic_prompt = f"""
You are an expert 'retrieval critic' assistant. Your task is to evaluate the relevance and utility of a set of retrieved document chunks in relation to a user's query. Your goal is to determine if these chunks provide sufficient and relevant information to answer the query accurately.

Instructions:
1.  Read the user's `Query` carefully.
2.  Review all `Retrieved Context Chunks` provided.
3.  Assess the overall relevance and utility of *all* the chunks combined to answer the `Query`.
4.  Based on your assessment, decide if the combined chunks are 'Suitable' (`YES` or `NO`) for generating an answer.
5.  Provide a brief `Explanation` for your decision.

Respond in the following JSON format:
```json
{{
  "Suitable": "YES/NO",
  "Explanation": "Your explanation here."
}}
```

Example:
Query: "What are the main causes of climate change?"
Retrieved Context Chunks: ["Human activities, such as the burning of fossil fuels, deforestation, and industrial processes, are the primary drivers of climate change.", "The greenhouse effect is a natural process that warms the Earth's surface."]

Output:
```json
{{
  "Suitable": "YES",
  "Explanation": "The chunks directly address the causes of climate change, mentioning human activities and the greenhouse effect."
}}
```

Query: "What is the capital of France?"
Retrieved Context Chunks: ["The Eiffel Tower is a famous landmark in Paris.", "Paris is known for its art museums."]

Output:
```json
{{
  "Suitable": "NO",
  "Explanation": "While the chunks mention Paris, they do not explicitly state that it is the capital of France, nor do they provide information to directly answer the query."
}}
```

Now, evaluate the following:

Query: {query}
Retrieved Context Chunks: {context_chunks_str}
"""

    try:
        response = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=[
                {"role": "system", "content": "You are a helpful assistant that evaluates retrieved context chunks."},
                {"role": "user", "content": retrieval_critic_prompt}
            ],
            response_format={"type": "json_object"},
            temperature=0.1
        )
        critic_output = json.loads(response.choices[0].message.content)

        suitable = critic_output.get("Suitable", "NO").upper()
        explanation = critic_output.get("Explanation", "No explanation provided.")

        print(f"**Retrieval Critic Decision:** {suitable} - {explanation}")

        if suitable == "YES":
            return retrieved_chunks
        else:
            return []
    except json.JSONDecodeError as e:
        print(f"Retrieval Critic Error: Could not parse LLM response as JSON: {e}. Response: {response.choices[0].message.content}")
        return []
    except Exception as e:
        print(f"Retrieval Critic Error: An unexpected error occurred: {e}")
        return []

print("Defined implement_retrieval_critic function.")
````

## Implement Generation Critic Function

````python
import json # Ensure json is imported for parsing LLM response

def implement_generation_critic(query: str, context_chunks: list, generated_answer: str):
    """
    Evaluates the faithfulness and relevancy of a generated answer using an LLM critic.

    Args:
        query (str): The user's query.
        context_chunks (list): A list of langchain Document objects that were used as context.
        generated_answer (str): The answer generated by the RAG system.

    Returns:
        dict: A dictionary containing the critic's assessment ('Faithful', 'Relevant', 'Explanation'),
              or an error dictionary if evaluation fails.
    """
    if not context_chunks:
        print("Generation Critic: No context chunks provided. Cannot evaluate faithfulness.")
        return {"Faithful": "NO", "Relevant": "NO", "Explanation": "No context provided."}

    # Prepare context for the prompt
    context_chunks_str = json.dumps([doc.page_content for doc in context_chunks], indent=2)

    generation_critic_prompt = f"""
You are an expert 'generation critic' assistant. Your task is to evaluate a generated answer based on a given user query and the context provided. You need to assess two key aspects: faithfulness and relevancy.

Instructions:
1.  **Faithfulness**: Determine if the `Generated Answer` is fully supported by the `Context Chunks`. Every statement in the generated answer must be directly inferable from the provided context. Mark as 'YES' if faithful, 'NO' otherwise.
2.  **Relevancy**: Determine if the `Generated Answer` directly and comprehensively addresses the `Query`. Mark as 'YES' if relevant, 'NO' otherwise.
3.  Provide a brief `Explanation` for your faithfulness and relevancy decisions.

Respond in the following JSON format:
```json
{{
  "Faithful": "YES/NO",
  "Relevant": "YES/NO",
  "Explanation": "Your explanation here."
}}
```

Example 1 (Good Answer):
Query: "What are the main causes of climate change?"
Context Chunks: ["Human activities, such as the burning of fossil fuels, deforestation, and industrial processes, are the primary drivers of climate change.", "The greenhouse effect is a natural process that warms the Earth's surface."]
Generated Answer: "The primary drivers of climate change are human activities like burning fossil fuels, deforestation, and industrial processes."

Output:
```json
{{
  "Faithful": "YES",
  "Relevant": "YES",
  "Explanation": "The answer is fully supported by the context and directly answers the question."
}}
```

Example 2 (Bad Answer - Not Faithful):
Query: "What is the capital of France?"
Context Chunks: ["The Eiffel Tower is a famous landmark in Paris.", "Paris is known for its art museums."]
Generated Answer: "The capital of France is Paris, a city famous for the Eiffel Tower and its art museums."

Output:
```json
{{
  "Faithful": "NO",
  "Relevant": "YES",
  "Explanation": "While the answer is relevant to the query, the context provided does not explicitly state that Paris is the capital of France, making the answer not fully faithful to the given context."
}}
```

Example 3 (Bad Answer - Not Relevant):
Query: "What are nitrate aerosols?"
Context Chunks: ["Nitrate aerosols are tiny particles in the atmosphere that contain nitrate compounds. They are significant because they can influence atmospheric chemistry.", "Sulfate aerosols are another type of atmospheric particle."]
Generated Answer: "Sulfate aerosols are important atmospheric particles."

Output:
```json
{{
  "Faithful": "YES",
  "Relevant": "NO",
  "Explanation": "The answer is faithful to a part of the context, but it does not directly answer the question about nitrate aerosols."
}}
```

Now, evaluate the following:
Query: {query}
Context Chunks: {context_chunks_str}
Generated Answer: {generated_answer}
"""

    try:
        response = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=[
                {"role": "system", "content": "You are a helpful assistant that evaluates generated answers."},
                {"role": "user", "content": generation_critic_prompt}
            ],
            response_format={"type": "json_object"},
            temperature=0.1
        )
        critic_output = json.loads(response.choices[0].message.content)

        faithful = critic_output.get("Faithful", "NO").upper()
        relevant = critic_output.get("Relevant", "NO").upper()
        explanation = critic_output.get("Explanation", "No explanation provided.")

        print(f"**Generation Critic Decision:** Faithful: {faithful}, Relevant: {relevant} - {explanation}")

        return {"Faithful": faithful, "Relevant": relevant, "Explanation": explanation}
    except json.JSONDecodeError as e:
        print(f"Generation Critic Error: Could not parse LLM response as JSON: {e}. Response: {response.choices[0].message.content}")
        return {"Faithful": "ERROR", "Relevant": "ERROR", "Explanation": f"JSON parsing error: {e}"}
    except Exception as e:
        print(f"Generation Critic Error: An unexpected error occurred: {e}")
        return {"Faithful": "ERROR", "Relevant": "ERROR", "Explanation": f"Unexpected error: {e}"}

print("Defined implement_generation_critic function.")
````

### 7. Building the RAG system

````python
def build_rag(documents: list):
    """
    Builds a Retrieval-Augmented Generation (RAG) system from a list of documents.

    Args:
        documents (list): A list of langchain Document objects.

    Returns:
        tuple: A tuple containing the FAISS index and the list of processed chunks.
    """
    print(f"Building RAG system with {len(documents)} initial documents...")

    if not documents:
        print("No documents provided to build the RAG system.")
        return None, []

    # 1. Clean each page (content of each Document)
    for doc in documents:
        doc.page_content = clean_text(doc.page_content)

    # 2. Split into chunks (Documents with metadata)
    chunks = split_docs(documents)

    # 3. Create embeddings for each chunk
    embeddings = create_embeddings(chunks)

    # 4. Build vector DB (FAISS) with chunks + embeddings
    index, processed_chunks = build_vector_db(chunks, embeddings)

    if index is None:
        print("RAG system build failed.")
        return None, []

    print(f"Ready! Created {len(processed_chunks)} chunks.")
    return index, processed_chunks
````

### 8. Chat loop function

````python
def chat(index, chunks):
    """
    Starts an interactive chat loop with the RAG system.

    Args:
        index: The FAISS index.
        chunks (list): The list of langchain Document objects corresponding to the index.
    """
    if index is None or not chunks:
        print("RAG system not properly initialized. Cannot start chat.")
        return

    print("Chat started. Type 'quit' to exit.")
    while True:
        query = input("Ask a question: ")
        if query.lower() == 'quit':
            break

        # Call the orchestrating generate_answer which handles search, retrieval critic, generation, and generation critic
        # This single call replaces the previous manual search, retrieval critic application, and answer generation.
        generated_answer_output, relevant_chunks_from_self_rag, generation_critic_result = generate_answer(query, index, chunks)

        # Check if the retrieval critic decided against generating an answer
        if generated_answer_output == "No suitable context found to answer the question.":
            print("\n**Retrieval Critic determined no suitable context was found. Cannot generate an answer for this query.**\n")
            continue

        # Format the answer to break lines after every 15 words
        words = generated_answer_output.split()
        formatted_answer_lines = []
        for i in range(0, len(words), 15):
            formatted_answer_lines.append(' '.join(words[i:i+15]))
        formatted_answer = '\n'.join(formatted_answer_lines)

        print(f"\n**Answer:**\n{formatted_answer}\n")

        print(f"**Generation Critic Report:** Faithful: {generation_critic_result['Faithful']}, Relevant: {generation_critic_result['Relevant']}")
        # Format the explanation to break lines after every 15 words
        explanation_words = generation_critic_result['Explanation'].split()
        formatted_explanation_lines = []
        for i in range(0, len(explanation_words), 15):
            formatted_explanation_lines.append(' '.join(explanation_words[i:i+15]))
        formatted_explanation = '\n'.join(formatted_explanation_lines)
        print(f"**Explanation:** {formatted_explanation}\n")

        # Show which chunks + metadata were used
        if relevant_chunks_from_self_rag:
            print("**Context provided to the model (top relevant chunks after critic):**")
            for i, doc in enumerate(relevant_chunks_from_self_rag, 1):
                meta = doc.metadata
                # Assuming 'page' is a common metadata key, adjust if needed.
                page_info = meta.get('page', 'unknown') if 'page' in meta else 'unknown'
                print(f"--- Chunk {i} (Source: {meta.get('source', 'unknown_source')}, Page: {page_info}):")
                print(doc.page_content[:300], "...\n")  # show first 300 chars for readability
        else:
            print("No specific context chunks were deemed suitable by the Retrieval Critic for this answer.")

        print('=' * 100)
````

### 9. Functions to Save and Load RAG System

````python
import pickle
import os

def save_rag_system(index, chunks, base_name: str = "rag_system"): # This function will now save to the current directory, if needed for local operations
    """
    Saves the FAISS index and associated chunks to disk.

    Args:
        index: The FAISS index to save.
        chunks (list): The list of langchain Document objects to save.
        base_name (str): The base name for the saved files (e.g., 'rag_system').
    """
    index_path = f"{base_name}.faiss"
    chunks_path = f"{base_name}_chunks.pkl"

    try:
        faiss.write_index(index, index_path)
        with open(chunks_path, "wb") as f:
            pickle.dump(chunks, f)
        print(f"RAG system saved: {index_path} and {chunks_path}")
    except Exception as e:
        print(f"Error saving RAG system: {e}")

def load_rag_system(base_name: str = "rag_system", drive_folder: str = "My_RAG_Systems"):
    """
    Loads the FAISS index and associated chunks from disk, specifically from Google Drive.

    Args:
        base_name (str): The base name for the saved files.
        drive_folder (str): The name of the folder in Google Drive's 'My Drive' where files are stored.

    Returns:
        tuple: A tuple containing the loaded FAISS index and the list of chunks,
               or (None, None) if loading fails.
    """
    drive_path = f'/content/drive/My Drive/{drive_folder}'
    index_path = os.path.join(drive_path, f"{base_name}.faiss")
    chunks_path = os.path.join(drive_path, f"{base_name}_chunks.pkl")

    if not os.path.exists(index_path) or not os.path.exists(chunks_path):
        print(f"Saved RAG system not found at {index_path} or {chunks_path}")
        return None, None

    try:
        index = faiss.read_index(index_path)
        with open(chunks_path, "rb") as f:
            chunks = pickle.load(f)
        print(f"RAG system loaded: {index_path} and {chunks_path}")
        return index, chunks
    except Exception as e:
        print(f"Error loading RAG system from {drive_path}: {e}")
        return None, None
````

## RAG System Evaluation

To assess the performance of our RAG system, we will implement several key evaluation metrics. Since `ragas` encountered dependency issues, we will define custom functions to calculate these metrics using direct LLM calls via the `openai` client. This approach allows us to control the evaluation process and understand each step.

### Metrics to be evaluated:

1.  **Context Relevancy**: How relevant are the retrieved document chunks to the user's query?
2.  **Answer Faithfulness**: Is the generated answer factually consistent with the retrieved context?
3.  **Answer Relevancy**: Does the generated answer directly and comprehensively address the user's question?
4.  **Context Recall (Simplified)**: Is the original document chunk (from which the evaluation question was derived) successfully retrieved by the RAG system?

### 1. Generating a Synthetic Evaluation Dataset

We need a set of questions, their ground-truth answers, the contexts retrieved by our RAG system, and the answers generated by our RAG system. We'll create this synthetically by taking existing document chunks, asking an LLM to generate a question and a concise answer *from that chunk*, and then running that question through our RAG system to get its output.

````python
import random
import json
import re # For parsing LLM responses

def generate_eval_dataset(chunks_list: list, num_samples: int = 10, rag_index=None, rag_chunks=None):
    """
    Generates a synthetic evaluation dataset for RAG systems.

    Args:
        chunks_list (list): A list of langchain Document objects (chunks) from which to generate questions.
        num_samples (int): The number of questions to generate.
        rag_index: The FAISS index of the RAG system.
        rag_chunks (list): The list of langchain Document objects corresponding to the RAG index.

    Returns:
        list: A list of dictionaries, each containing 'question', 'ground_truth_answer', 'contexts', 'answer', and 'original_chunk_content'.
    """
    eval_data = []
    print(f"Generating {num_samples} evaluation samples...")

    if rag_index is None or rag_chunks is None:
        print("Error: RAG index and chunks must be provided for evaluation dataset generation.")
        return []

    # Ensure there are enough chunks to sample
    if len(chunks_list) < num_samples:
        num_samples = len(chunks_list)
        print(f"Warning: Not enough chunks for {num_samples} samples, reducing to {len(chunks_list)}.")

    # Randomly sample chunks to create questions from
    sampled_chunks = random.sample(chunks_list, num_samples)

    for i, chunk in enumerate(sampled_chunks):
        print(f"Processing sample {i+1}/{num_samples}...")
        context_text = chunk.page_content

        # Use LLM to generate question and ground truth answer from the chunk
        qa_response = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=[
                {"role": "system", "content": "You are a helpful assistant. Generate a single question and a concise answer based ONLY on the provided text. The question should be answerable directly from the text. Respond in JSON format: { \"question\": \"your question\", \"answer\": \"your answer\" }"},
                {"role": "user", "content": f"Text: {context_text}"}
            ],
            response_format={"type": "json_object"},
            temperature=0.7
        )
        try:
            qa_pair = json.loads(qa_response.choices[0].message.content)
            question = qa_pair['question']
            ground_truth_answer = qa_pair['answer']
        except json.JSONDecodeError as e:
            print(f"Error decoding JSON from LLM for QA pair: {e}. Skipping sample.")
            continue

        # Use our Self-RAG system (via refactored generate_answer) to get generated answer and relevant contexts
        generated_answer_output, relevant_chunks_for_eval, generation_critic_output = generate_answer(question, rag_index, rag_chunks)

        eval_data.append({
            "question": question,
            "ground_truth_answer": ground_truth_answer,
            "contexts": [doc.page_content for doc in relevant_chunks_for_eval], # Use relevant_chunks_for_eval from Self-RAG
            "answer": generated_answer_output, # Use the answer from Self-RAG
            "original_chunk_content": context_text, # Store original chunk content for context recall
            "generation_critic": generation_critic_output # Store critic output
        })
    print("Evaluation dataset generation complete!")
    return eval_data

# --- Self-RAG Evaluation Dataset Generation ---
print("--- Attempting to load previously saved RAG system for Self-RAG Evaluation ---")
all_pdfs_index, all_pdfs_chunks = load_rag_system(base_name="full_rag_system")

if all_pdfs_index is None or not all_pdfs_chunks:
    print("Failed to load RAG system. Cannot generate evaluation samples for Self-RAG.")
else:
    print("--- Generating Evaluation Samples for Self-RAG ---")
    # Generate the evaluation dataset using the updated RAG system with critics
    # Increase `num_samples` for a more robust evaluation in a real scenario.
    self_rag_eval_dataset = generate_eval_dataset(all_pdfs_chunks, num_samples=5, rag_index=all_pdfs_index, rag_chunks=all_pdfs_chunks)

    print("\nGenerated Self-RAG Evaluation Samples:")
    for i, sample in enumerate(self_rag_eval_dataset):
        print(f"--- Sample {i+1} ---")
        print(f"Q: {sample['question']}")
        print(f"GT Answer: {sample['ground_truth_answer']}")
        print(f"RAG Answer: {sample['answer']}")
        # Safely print retrieved contexts
        if sample['contexts']:
            print(f"Retrieved Contexts (first chunk): {sample['contexts'][0][:200]}...")
        else:
            print("Retrieved Contexts: No suitable context found.")
        print(f"Generation Critic: Faithful: {sample['generation_critic']['Faithful']}, Relevant: {sample['generation_critic']['Relevant']}")
````

### 2. Custom Metric Calculation Functions

Here we define functions to calculate each of the desired evaluation metrics using LLM calls.

````python
def llm_score_evaluation(prompt: str, model: str = "gpt-4o-mini") -> str:
    """
    Helper function to get LLM response for evaluation scoring.
    """
    response = client.chat.completions.create(
        model=model,
        messages=[
            {"role": "user", "content": prompt + "\nRespond with only the rating number."}
        ],
        temperature=0.0 # Make it deterministic for scoring
    )
    return response.choices[0].message.content


def calculate_context_relevancy(eval_data: list) -> float:
    """
    Calculates the average context relevancy score using an LLM.
    Scores each retrieved chunk's relevancy to the question on a scale of 1 to 5.
    """
    scores = []
    print("Calculating Context Relevancy...")
    for i, item in enumerate(eval_data):
        question = item["question"]
        contexts = item["contexts"]
        if not contexts:
            scores.append(1) # If no context, it's not relevant to anything, lowest score
            continue

        chunk_scores = []
        for context_chunk in contexts:
            prompt = f"""
            Given the following question and context, rate the context's relevance to the question on a scale of 1 to 5.
            1: Not relevant at all
            2: Slightly relevant
            3: Moderately relevant
            4: Highly relevant
            5: Directly answers the question or provides essential information

            Question: {question}
            Context: {context_chunk}
            """
            response = llm_score_evaluation(prompt)
            try:
                rating_match = re.search(r'(\d+)', response) # Modified regex to extract just the number
                if rating_match:
                    rating = int(rating_match.group(1))
                    chunk_scores.append(rating)
                else:
                    print(f"Warning: Could not parse rating for context relevancy in sample {i}. Response: {response}. Defaulting to 1.")
                    chunk_scores.append(1) # Default to lowest if parsing fails
            except (AttributeError, ValueError):
                print(f"Error parsing rating for context relevancy in sample {i}. Response: {response}. Defaulting to 1.")
                chunk_scores.append(1) # Default to lowest if parsing fails

        if chunk_scores:
            scores.append(sum(chunk_scores) / len(chunk_scores))
        else:
            scores.append(1) # Default if no valid chunk scores were obtained

    return sum(scores) / len(scores) if scores else 0

def calculate_answer_faithfulness(eval_data: list) -> float:
    """
    Calculates the average answer faithfulness score using an LLM.
    Determines if the generated answer is fully supported by the provided context (binary score).
    """
    scores = []
    print("Calculating Answer Faithfulness...")
    for i, item in enumerate(eval_data):
        answer = item["answer"]
        contexts = item["contexts"]
        if not contexts:
            scores.append(0) # Cannot be faithful if no context
            continue

        context_str = "\n\n".join(contexts)

        prompt = f"""
        Given the following generated answer and context, determine if the generated answer is fully supported by the provided context.
        Respond with 'YES' if every statement in the answer can be directly inferred from the context. Respond with 'NO' otherwise (if it contains information not found in the context or contradicts the context).

        Generated Answer: {answer}
        Context: {context_str}
        """
        response = llm_score_evaluation(prompt)
        if "yes" in response.lower():
            scores.append(1)
        else:
            scores.append(0)
    return sum(scores) / len(scores) if scores else 0

def calculate_answer_relevancy(eval_data: list) -> float:
    """
    Calculates the average answer relevancy score using an LLM.
    Rates how relevant the generated answer is to the question on a scale of 1 to 5.
    """
    scores = []
    print("Calculating Answer Relevancy...")
    for i, item in enumerate(eval_data):
        question = item["question"]
        answer = item["answer"]

        prompt = f"""
        Given the following question and answer, rate how relevant the answer is to the question on a scale of 1 to 5.
        1: Not relevant
        2: Slightly relevant
        3: Moderately relevant
        4: Highly relevant
        5: Directly and comprehensively answers the question

        Question: {question}
        Answer: {answer}
        """
        response = llm_score_evaluation(prompt)
        try:
            rating_match = re.search(r'(\d+)', response) # Modified regex to extract just the number
            if rating_match:
                rating = int(rating_match.group(1))
                scores.append(rating)
            else:
                print(f"Warning: Could not parse rating for answer relevancy in sample {i}. Response: {response}. Defaulting to 1.")
                scores.append(1) # Default to lowest if parsing fails
        except (AttributeError, ValueError):
            print(f"Error parsing rating for answer relevancy in sample {i}. Response: {response}. Defaulting to 1.")
            scores.append(1) # Default to lowest if parsing fails
    return sum(scores) / len(scores) if scores else 0

def calculate_context_recall(eval_data: list) -> float:
    """
    Calculates the context recall (simplified) by checking if the original chunk content
    (from which the question was generated) is present in the retrieved contexts.
    """
    hits = 0
    print("Calculating Context Recall...")
    for item in eval_data:
        original_chunk_content = item["original_chunk_content"]
        retrieved_contexts = item["contexts"]

        # Check if the content of the original chunk is present in any of the retrieved contexts
        # A simple substring check is used here for simplicity. For more robustness,
        # one might consider embedding similarity or more advanced text comparison.
        if any(original_chunk_content in rc for rc in retrieved_contexts):
            hits += 1
    return hits / len(eval_data) if eval_data else 0
````

### 3. Run Evaluation and Display Results

Now we will execute the evaluation functions and display the computed metrics.

````python
print("--- Starting RAG System Evaluation ---")

# Ensure 'self_rag_eval_dataset' is populated from the previous step
if not self_rag_eval_dataset:
    print("Evaluation dataset is empty. Please generate it first.")
else:
    # Calculate metrics
    context_relevancy_score = calculate_context_relevancy(self_rag_eval_dataset)
    answer_faithfulness_score = calculate_answer_faithfulness(self_rag_eval_dataset)
    answer_relevancy_score = calculate_answer_relevancy(self_rag_eval_dataset)
    context_recall_score = calculate_context_recall(self_rag_eval_dataset)

    print("\n--- RAG Evaluation Results ---")
    print(f"Average Context Relevancy: {context_relevancy_score:.2f} (Scale: 1-5)")
    print(f"Average Answer Faithfulness: {answer_faithfulness_score:.2f} (Binary: 0=No, 1=Yes)")
    print(f"Average Answer Relevancy: {answer_relevancy_score:.2f} (Scale: 1-5)")
    print(f"Average Context Recall: {context_recall_score:.2f} (Binary: 0=No, 1=Yes)")

    print("\nEvaluation complete!")
````

## Self-RAG System Flow Overview

This section provides a conceptual overview of how the Self-RAG system operates, integrating both the Retrieval Critic and Generation Critic to enhance the quality and reliability of generated answers.

```
+-----------------------+
|     User Query        |
+----------+------------+
           |
           v
+----------+------------+
|   Search Chunks        |
| (Initial Retrieval)    |
+----------+------------+
           |
           v
+----------+------------+
| Retrieval Critic (LLM) |
| (Checks chunk relevance)|
+----------+------------+
     |      YES         NO
     | (Chunks are       |
     v   Relevant)       |
+----------+------------+   |
| Generate Answer        |   |
| (using Relevant Chunks)|   |
+----------+------------+   |
           |                   |
           v                   v
+----------+------------+   +------------------------------------+
| Generation Critic (LLM)|   | Report: "No suitable context found"|
| (Checks faithfulness &  |   | (Prevents hallucination)           |
| relevancy of answer)   |   +------------------------------------+
+----------+------------+
           |
           v
+----------+------------+
|  Formatted Output      |
| (Answer + Critic Report)|
+-----------------------+

```

### **Detailed Flow:**

1.  **User Query:** The process begins when the user submits a question to the RAG system.

2.  **Initial Retrieval (`search_chunks`):** Based on the user's query, the system searches the FAISS vector database to retrieve a set of potentially relevant document chunks. These chunks are considered initial candidates.

3.  **Retrieval Critic (`implement_retrieval_critic`):**
    *   The user's query and the `initial retrieved chunks` are passed to an **LLM-based Retrieval Critic**.
    *   The critic evaluates whether the combined `retrieved chunks` contain sufficient and relevant information to answer the `query` accurately.
    *   **Decision:**
        *   If the critic deems the chunks `YES` (suitable), these chunks are passed on for answer generation.
        *   If the critic deems the chunks `NO` (not suitable), the system stops here and reports to the user that **"No suitable context was found. Cannot generate an answer for this query."** This is a critical step in preventing hallucinations, as it ensures answers are only generated from validated contexts.

4.  **Generate Answer (`_raw_generate_answer`):**
    *   If the Retrieval Critic approves the chunks, the `query` and the `critic-approved relevant chunks` are sent to the core generation LLM (`_raw_generate_answer`).
    *   This LLM then synthesizes an answer based *only* on the provided `relevant chunks`.

5.  **Generation Critic (`implement_generation_critic`):**
    *   The original `query`, the `critic-approved relevant chunks`, and the `generated answer` are all passed to an **LLM-based Generation Critic**.
    *   This critic evaluates two key aspects of the `generated answer`:
        *   **Faithfulness:** Is the answer fully supported by the provided `context chunks`?
        *   **Relevancy:** Does the answer directly and comprehensively address the `query`?
    *   The critic provides a `Faithful` score (`YES`/`NO`), a `Relevant` score (`YES`/`NO`), and an `Explanation` for its assessment.

6.  **Formatted Output:**
    *   Finally, the system presents the `generated answer` to the user.
    *   Crucially, it also includes the `Generation Critic's report` (Faithful, Relevant, and Explanation) alongside the answer. This provides transparency and allows the user to gauge the confidence and quality of the response.
    *   The `relevant chunks` (after the Retrieval Critic's assessment) are also shown, along with their source and page information, allowing the user to verify the context directly.

This iterative process of critical assessment at both the retrieval and generation stages makes the Self-RAG system more robust, reliable, and transparent compared to a traditional RAG system.
